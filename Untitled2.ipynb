```python
import streamlit as st
import streamlit.components.v1 as components
import pandas as pd
import networkx as nx
from pyvis.network import Network
import numpy as np
import pandas as pd
df_nodes = pd.read_csv('data/nodes.csv')
df_edges = pd.read_csv('data/edges.csv')
df_edges.head(10)
```

```python
g = Network()
g.add_nodes(df_nodes['index'].tolist(), label=df_nodes['node'].tolist(), color=df_nodes['color'].tolist())

a = df_edges['agency'].tolist()
b = df_edges['contractor'].tolist()
c = df_edges['adj_weight'].tolist()

for i in range(len(a)):
    g.add_edge(a[i], b[i],weight=c[i])
# df_nodes['node'].tolist()
```
